```python
from manim import *

config.media_width = "80%"
```

```python
%%manim -v WARNING -qm ProyeccionesYSimetrias

class ProyeccionesYSimetrias(Scene):
    
    def parpadear(self, vector, nombre_de_vector, color_de_vector, elemento, color_de_calculo):
        self.play(
            FadeToColor(nombre_de_vector, YELLOW),
            FadeToColor(vector, YELLOW),
            FadeToColor(elemento, YELLOW),
            runtime=.5
        )
        self.wait(.5)
        self.play(
            FadeToColor(nombre_de_vector, color_de_vector),
            FadeToColor(vector, color_de_vector),
            FadeToColor(elemento, color_de_calculo),
            runtime=.5
        )
        self.wait(.5)
    
    def resaltar_suma_de_vectores(self, 
            vector_suma, nombre_de_vector_suma, color_de_vector_suma, 
            vector_resta, nombre_de_vector_resta, color_de_vector_resta, 
            elemento_suma, elemento_resta, color_calculo
        ):
        self.parpadear(vector_suma, nombre_de_vector_suma, color_de_vector_suma, elemento_suma, color_calculo)
        self.parpadear(vector_resta, nombre_de_vector_resta, color_de_vector_resta, elemento_resta, color_calculo)
    
    def construct(self):

        plane = NumberPlane()

        vector = Line(
            start=plane.coords_to_point(0, 0),
            end=plane.coords_to_point(5, 3),
            stroke_color=YELLOW,
        ).add_tip()
        nombre_de_vector = (
            MathTex("v").next_to(vector, UP, buff=0).set_color(YELLOW)
        )
        
        calculo_vector_con_proyecciones = (
            MathTex("v = \Pi_{\mathcal{S}_{1}\mathcal{S}_{2}}(v) + \Pi_{\mathcal{S}_{2}\mathcal{S}_{1}}(v)").next_to(vector, UP, buff=0).set_color(YELLOW)
        )
        
        proyeccion_de_v_respecto_a_s1_en_direccion_de_s2 = Line(
            start=plane.coords_to_point(0, 0),
            end=plane.coords_to_point(5, 0),
            stroke_color=GREEN,
        ).add_tip()
        proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_nombre = (
            MathTex("\Pi_{\mathcal{S}_{1}\mathcal{S}_{2}}(v)").next_to(proyeccion_de_v_respecto_a_s1_en_direccion_de_s2, DOWN, buff=0).set_color(GREEN)
        )
        
        proyeccion_de_v_respecto_a_s2_en_direccion_de_s1 = Line(
            start=plane.coords_to_point(0, 0),
            end=plane.coords_to_point(0, 3),
            stroke_color=RED,
        ).add_tip()
        proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_nombre = (
            MathTex("\Pi_{\mathcal{S}_{2}\mathcal{S}_{1}}(v)").next_to(proyeccion_de_v_respecto_a_s2_en_direccion_de_s1, LEFT, buff=0).set_color(RED)
        )
        
        proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_negativo = Line(
            start=plane.coords_to_point(0, 3),
            end=plane.coords_to_point(-5, 3),
            stroke_color=GREEN,
        ).add_tip()
        proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_negativo_nombre = (
            MathTex("\Pi_{\mathcal{S}_{1}\mathcal{S}_{2}}(v)").next_to(proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_negativo, UP, buff=0).set_color(GREEN)
        )
        
        proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_negativo = Line(
            start=plane.coords_to_point(5, 0),
            end=plane.coords_to_point(5, -3),
            stroke_color=RED,
        ).add_tip()
        proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_negativo_nombre = (
            MathTex("\Pi_{\mathcal{S}_{2}\mathcal{S}_{1}}(v)").next_to(proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_negativo, RIGHT, buff=0).set_color(RED)
        )
        
        simetria_de_v_respecto_a_s1_en_direccion_de_s2 = Line(
            start=plane.coords_to_point(0, 0),
            end=plane.coords_to_point(5, -3),
            stroke_color=WHITE,
        ).add_tip()
        simetria_de_v_respecto_a_s1_en_direccion_de_s2_nombre = (
            MathTex("\Sigma_{\mathcal{S}_{1}\mathcal{S}_{2}}(v)").next_to(simetria_de_v_respecto_a_s1_en_direccion_de_s2, DOWN, buff=0).set_color(WHITE)
        )
        
        calculo_simetria_de_v_respecto_a_s1_en_direccion_de_s2 = (
            MathTex("\Sigma_{\mathcal{S}_{1}\mathcal{S}_{2}}(v) = ","\Pi_{\mathcal{S}_{1}\mathcal{S}_{2}}(v)"," - \Pi_{\mathcal{S}_{2}\mathcal{S}_{1}}(v)").next_to(simetria_de_v_respecto_a_s1_en_direccion_de_s2, DOWN, buff=0).set_color(WHITE)
        )
        
        simetria_de_v_respecto_a_s2_en_direccion_de_s1 = Line(
            start=plane.coords_to_point(0, 0),
            end=plane.coords_to_point(-5, 3),
            stroke_color=WHITE,
        ).add_tip()
        simetria_de_v_respecto_a_s2_en_direccion_de_s1_nombre = (
            MathTex("\Sigma_{\mathcal{S}_{2}\mathcal{S}_{1}}(v)").next_to(simetria_de_v_respecto_a_s2_en_direccion_de_s1, DOWN, buff=0).set_color(WHITE)
        )
        
        calculo_simetria_de_v_respecto_a_s2_en_direccion_de_s1 = (
            MathTex("\Sigma_{\mathcal{S}_{2}\mathcal{S}_{1}}(v) = ","\Pi_{\mathcal{S}_{2}\mathcal{S}_{1}}(v)"," - \Pi_{\mathcal{S}_{1}\mathcal{S}_{2}}(v)").next_to(simetria_de_v_respecto_a_s2_en_direccion_de_s1, DOWN, buff=0).set_color(WHITE)
        )
        
        stuff = VGroup(
            plane, 
            vector, nombre_de_vector, 
            proyeccion_de_v_respecto_a_s1_en_direccion_de_s2, proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_nombre,
            proyeccion_de_v_respecto_a_s2_en_direccion_de_s1, proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_nombre,
            proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_negativo, proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_negativo_nombre,
            proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_negativo, proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_negativo_nombre,
            simetria_de_v_respecto_a_s1_en_direccion_de_s2, simetria_de_v_respecto_a_s1_en_direccion_de_s2_nombre,
            simetria_de_v_respecto_a_s2_en_direccion_de_s1, simetria_de_v_respecto_a_s2_en_direccion_de_s1_nombre,
        )

        self.play(
            Write(plane),
            Write(MathTex("\mathcal{S}_{2}").to_edge(UP, buff=0)),
            Write(MathTex("\mathcal{S}_{1}").to_edge(RIGHT, buff=0)),
        )
        self.wait()
        self.play(
            GrowFromPoint(vector, point=vector.get_start()), 
            Write(nombre_de_vector), 
        )
        self.wait()
        self.play(
            ReplacementTransform(vector.copy(), proyeccion_de_v_respecto_a_s1_en_direccion_de_s2),
            Write(proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_nombre), 
            ReplacementTransform(vector.copy(), proyeccion_de_v_respecto_a_s2_en_direccion_de_s1),
            Write(proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_nombre), 
        )
        self.wait()
        
        copia_de_nombre_de_vector = nombre_de_vector.copy()
        self.play(
            ReplacementTransform(nombre_de_vector, calculo_vector_con_proyecciones),
        )
        self.wait()
        self.play(
            ReplacementTransform(calculo_vector_con_proyecciones, copia_de_nombre_de_vector),
        )
        self.wait()

        self.play(
            ReplacementTransform(vector.copy(), simetria_de_v_respecto_a_s2_en_direccion_de_s1), 
            Write(simetria_de_v_respecto_a_s2_en_direccion_de_s1_nombre), 
        )
        self.wait()
        
        simetria_de_v_respecto_a_s2_en_direccion_de_s1_nombre_copia = simetria_de_v_respecto_a_s2_en_direccion_de_s1_nombre.copy()
        proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_copia = proyeccion_de_v_respecto_a_s1_en_direccion_de_s2.copy()
        proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_nombre_copia = proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_nombre.copy()
        self.play(
            Transform(proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_copia, proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_negativo),
            Transform(proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_nombre_copia, proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_negativo_nombre),
            ReplacementTransform(simetria_de_v_respecto_a_s2_en_direccion_de_s1_nombre, calculo_simetria_de_v_respecto_a_s2_en_direccion_de_s1),
        )
        self.wait()
        
        self.resaltar_suma_de_vectores(
            proyeccion_de_v_respecto_a_s2_en_direccion_de_s1, proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_nombre, RED,
            proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_copia, proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_nombre_copia, GREEN,
            calculo_simetria_de_v_respecto_a_s2_en_direccion_de_s1[1], calculo_simetria_de_v_respecto_a_s2_en_direccion_de_s1[2], WHITE
        )
    
        self.play(
            ReplacementTransform(calculo_simetria_de_v_respecto_a_s2_en_direccion_de_s1, simetria_de_v_respecto_a_s2_en_direccion_de_s1_nombre_copia),
            FadeOut(proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_nombre_copia),
            FadeOut(proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_copia)
        )
        self.wait()
        
        self.play(
            ReplacementTransform(vector.copy(), simetria_de_v_respecto_a_s1_en_direccion_de_s2), 
            Write(simetria_de_v_respecto_a_s1_en_direccion_de_s2_nombre), 
        )
        self.wait()
        
        simetria_de_v_respecto_a_s1_en_direccion_de_s2_nombre_copia = simetria_de_v_respecto_a_s1_en_direccion_de_s2_nombre.copy()
        proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_copia = proyeccion_de_v_respecto_a_s2_en_direccion_de_s1.copy()
        proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_nombre_copia = proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_nombre.copy()
        self.play(
            Transform(proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_copia, proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_negativo),
            Transform(proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_nombre_copia, proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_negativo_nombre),
            ReplacementTransform(simetria_de_v_respecto_a_s1_en_direccion_de_s2_nombre, calculo_simetria_de_v_respecto_a_s1_en_direccion_de_s2),
        )
        self.wait()
        
        self.resaltar_suma_de_vectores(
            proyeccion_de_v_respecto_a_s1_en_direccion_de_s2, proyeccion_de_v_respecto_a_s1_en_direccion_de_s2_nombre, GREEN,
            proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_copia, proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_nombre_copia, RED,
            calculo_simetria_de_v_respecto_a_s1_en_direccion_de_s2[1], calculo_simetria_de_v_respecto_a_s1_en_direccion_de_s2[2], WHITE
        )
    
        self.play(
            ReplacementTransform(calculo_simetria_de_v_respecto_a_s1_en_direccion_de_s2, simetria_de_v_respecto_a_s1_en_direccion_de_s2_nombre_copia),
            FadeOut(proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_nombre_copia),
            FadeOut(proyeccion_de_v_respecto_a_s2_en_direccion_de_s1_copia)
        )
        self.wait()
```
